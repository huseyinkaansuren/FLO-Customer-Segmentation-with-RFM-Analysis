##Step 8: Functionalize the data preparation process.

```python
def data_prep(dataframe):
  #1. Creating New Columns
  dataframe["total_order_num"] = dataframe["order_num_total_ever_online"] + dataframe["order_num_total_ever_offline"]
  dataframe["total_value"] = dataframe["customer_value_total_ever_offline"] + dataframe["customer_value_total_ever_online"]


  #2. Object to Date
  date_cols = [col for col in dataframe.columns if "date" in col]
  for col in date_cols:
    dataframe[col] = pd.to_datetime(dataframe[col].str.strip(), format= "%Y-%m-%d")


  return dataframe
```

```python
data_prep(df)
```

#Task 2: Calculating RFM Metrics

##Step 1: Define Recency, Frequency and Monetary.

```python
#Recency: It is the day difference between the date of the customer's last transaction and the analysis date.
#Frequency: Total transaction number of customer's
#Monetary: It is the total monetary value brought by the customer
```

##Step 2: Calculate Recency, Frequency and Monetary metrics for the customer.

```python
df["last_order_date"].max()
```

```python
today_date = dt.datetime(2021, 6, 1)
```

```python
rfm = df.groupby("master_id").agg({"last_order_date": lambda date: (today_date - date.max()).days,
                                   "total_order_num": lambda order_num: order_num,
                                   "total_value": lambda totalvalue: totalvalue})
```

```python
rfm.head()
```

```python
rfm.columns = ["recency", "frequency", "monetary"]
```

```python
rfm.head()
```

#Task 3: Calculating the RF Score

##Step 1: Convert Recency, Frequency and Monetary metrics into scores between 1-5 with the help of qcut. Step 2: Save these scores as a new variable by adding _score.

```python
rfm["recency_score"] = pd.qcut(rfm["recency"], 5, labels = [5,4,3,2,1])
```

```python
rfm["frequency_score"] = pd.qcut(rfm["frequency"].rank(method="first"), 5, labels=[1,2,3,4,5])
```

```python
rfm["monetary_score"] = pd.qcut(rfm["monetary"], 5, labels = [1,2,3,4,5])
```

```python
rfm.head()
```

##Step 3: Express recency_score and frequency_score as a single variable and save it as RF_Score.

```python
rfm["rf_score"] = rfm["recency_score"].astype(str) + rfm["frequency_score"].astype(str)
```

```python
rfm.head()
```

#Task 4: Defining RF Score as a Segment

##Step 1: Make segment definitions for the generated RF scores

```python
seg_map = {
    r"[1-2][1-2]": "hibernating",
    r"[1-2][3-4]": "at_Risk",
    r"[1-2]5": "cant_loose",
    r"3[1-2]": "about_to_sleep",
    r"33": "need_attention",
    r"[3-4][4-5]": "loyal_customers",
    r"41": "promising",
    r"51": "new_customers",
    r"[4-5][2-3]": "potential_loyalists",
    r"5[4-5]": "champions"
}
```

```python
rfm["segment"] = rfm["rf_score"].replace(seg_map, regex = True)
```

```python
rfm.head()
```
